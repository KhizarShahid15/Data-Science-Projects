<b><h2>Data Filteration</h2></b>
The data can be manipulated, filtered, cropped via a lot of different ways.

```python
#Values in Defense column that are greater than 210 are stored in a new DataFrame x.

x = data['Defense']>210
```

```python
data[x]
```

```python
#Using Logical AND operator to view the desired data only.

data[np.logical_and(data['Defense']>210, data['Attack']<140)]
```

```python
# Logical AND operatorn can also be used as follows.

data[(data['Defense']>210) & (data['Attack']<140)]
```

```python
# Applying the simple formula of calculating the average of all the values in column Speed.

average_speed = sum(data['Speed'])/len(data['Speed'])
```

The following command adds a new column in the data and the values of the column will be 'high' or 'low' depending on the average speed.

```python
# Using if-else and for loop in a single command.
# The command is also adding a new column in the data along with filling values in it.

data["Speed Level"] = ["high" if i>average_speed else "low" for i in data.Speed]
```

```python
# .loc locates the desired data values.
# The command shows the values of Speed and Speed Level columns,
# starting from the beginning till index 10. That makes 11 rows in total.

data.loc[:10,["Speed", "Speed Level"]]
```

```python
# To count all the values of any particular column, the following command can be used.
# dropna=False counts all the NAN values as well.

data["Type 2"].value_counts(dropna=False)
```

As can be seen above, there are 386 NAN values in Type 2 Column/Feature.

```python
data_1 = data
```

```python
# Dropping all the NAN values in a particular column.

data_1["Type 2"].dropna(inplace = True)
```

Checking via assert if there are any null values present

```python
# To check if there is any NAN value left.
# The output of this statement returns nothing which means that it is true that there are no NAN values left.

assert  data_1['Type 2'].notnull().all()
```

Running above command shows nothing which means that are no NAN values available anymore. Now fill NA with empty.

```python
# Filling NAN values with empty in a particular column.

data_1["Type 2"].fillna('empty', inplace=True)
```

```python
# To check if there is any NAN value left.
# The output of this statement returns nothing which means that it is true that there are no NAN values left.

assert  data_1['Type 2'].notnull().all()
```

```python
# To count all the values of any particular column, the following command can be used.
# dropna=False counts all the NAN values as well.

data_1['Type 2'].value_counts(dropna=False)
```

<b><h2>Indexing Pandas Time Series</h2></b>

Parsing dates from string to ISO 8601 (yyyy-mm-dd hh:mm:ss ) format

```python
# Declaring and Initializing a list of dates.

list_time = ["2017-01-05","2017-02-10"]
```

```python
# Printing out the stored list.

list_time
```

```python
# Checking out the data type of the list which is string in this case.

print(type(list_time[1]))
```

```python
# Changing the dates data type into ISO 8601 (yyyy-mm-dd hh:mm:ss) format.

date_time = pd.to_datetime(list_time)
```

```python
# Printing out the stored list.

date_time
```

```python
# Checking out the data type of the list.

print(type(date_time))
```

Let us do the same on the Pokemon data.

```python
# Storing the first five rows of the dataset in a separate dataframe.
data_time_indexing = data.head()

# Creating a list of dates for all the five rows.
list_dates = ["2017-01-05","2017-02-05","2017-03-05","2018-03-10","2018-03-11"]

# Converting the list of dates from data type string to ISO 8601 format.
dates_time = pd.to_datetime(list_dates)

# Adding a new column in the data and storing the dates in it.
data_time_indexing["Date"] = dates_time
```

```python
data_time_indexing
```

Setting date as index.

```python
# Setting dates as the index of the data set.

data_time_indexing = data_time_indexing.set_index("Date")
```

```python
data_time_indexing
```

Selecting data according to the index Date.

```python
# Selecting data using a single date as index.

print(data_time_indexing.loc["2018-03-11"])
```

```python
# Selecting data using a range of dates as index.

print(data_time_indexing.loc["2017-03-05":"2018-03-11"])
```

<b><h2>Resampling Pandas Time Series</h2></b>

Resampling, Downsampling, Upsampling and Interpolating the data.

```python
# Resampling data according on yearly basis. (A is denoted for year.)

data_time_indexing.resample("A").mean()
```

```python
# Resampling data according on monthly basis. (M is denoted for month.)
# The months for which there is no data shows NAN in their rows.

data_time_indexing.resample("M").mean()
```

```python
# Interpolating data using linear method from first value.

data_time_indexing.resample("M").first().interpolate("linear")
```

```python
# Interpolating data using linear method from mean.

data_time_indexing.resample("M").mean().interpolate("linear")
```
